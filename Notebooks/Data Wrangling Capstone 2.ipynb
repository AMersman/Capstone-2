```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import datetime

```

```python
artists_data = pd.read_csv('../data/artists.csv')
artists_data.info()
```

```python
tracks_data = pd.read_csv('../data/tracks.csv')
tracks_data.info()
```

```python
tracks_data.head(10)
```

```python
pd.to_datetime(tracks_data['release_date'])
tracks_data.info()
```

```python
year=pd.DatetimeIndex(tracks_data['release_date']).year
month=pd.DatetimeIndex(tracks_data['release_date']).month
tracks_data.insert(8, 'month', month)
tracks_data.insert(9, 'year', year)

```

```python
tracks_data.sort_values('popularity', ascending =False).head(10)
```

```python
tracks_data['name'].value_counts().head(10)
```

```python
tracks_data[tracks_data['name']=='Summertime'].sort_values('year')
```

```python
column_names = ['name', 'month', 'year', 'release_date', 'artists', 'id_artists']
summaries = {'popularity': 'max', 'loudness': 'mean', 'speechiness':'mean', 'acousticness':'mean', 'instrumentalness':'mean', 'liveness':'mean', 'valence':'mean', 'tempo':'mean', 'duration_ms':'mean'}
tracks_data_clean = tracks_data.groupby(column_names).agg(summaries).reset_index()

tracks_data_clean['name'].value_counts()
```

```python
duplicates = tracks_data_clean.duplicated(column_names, keep = False)
tracks_data_clean[duplicates].sort_values('name')
```

```python
tracks_data_clean[tracks_data_clean['name']=='Summertime'].sort_values('year')
```
